## Task 2: MatPlotLib Basics

### 2a) Plot the data in a single plot

Using the [MatPlotLib](https://matplotlib.org/) documentation, visualize the dataset with Matplotlib. All charts should include a title, axis labels and a legend when applicable.

1. Import 'lab1.csv' into a ndarray using genfromtxt. 
2. Set figure size to 12x8. [Figure reference](https://matplotlib.org/3.1.0/api/_as_gen/matplotlib.pyplot.figure.html) [rcParams](https://matplotlib.org/3.1.0/tutorials/introductory/customizing.html#matplotlib-rcparams)
2. Create a cartesian line graph of column 2 (on the X axis) vs column 1 (Y axis) using a red dashed line. [matplotlib.pyplot introduction](https://matplotlib.org/tutorials/introductory/pyplot.html)
    1. Include a title and axis labels.

```python
import matplotlib.pyplot as plt
import numpy as np
lab1 = np.genfromtxt(fname = '/Users/adriangellert/Documents/Colby_2223/Spring/CS251/Lab01b/data/lab1.csv', delimiter= ",", skip_header = 1)
col2 = lab1[:,2]
col1 = lab1[:,1]

plt.rcParams['lines.color'] = 'r'
plot = plt.figure(figsize = [12,8])
plot = plt.plot(col2,col1, color = 'red', ls = 'dashed')
plot = plt.xlabel('column 2 data')
plot = plt.ylabel('column 1 data')
plot = plt.title('line graph of column 2 vs column 1')
```

### 2b) Plot the data in multiple subplots

Create 6 graphs in 2 columns and 3 rows using [subplots](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.subplot.html). 
1. Row 1 is two subplots (a scatterplot and a line graph) of column 1 graphed on the x-axis VS column 0 graphed on the y-axis

```python
fig1 = plt.subplot(3, 2, 1) # three rows, two columns, first figure
col0 = lab1[:,0]
# col1 created previously
# col2 created previously
# plot1 = plt.scatter(col1,col0, color = 'red')
plt.scatter(col1,col0)
plt.xlabel('col1 data')
plt.ylabel('col2 data')
plt.title('scatter plot of col1 vs col0')

plt.subplot(3, 2, 2)
plot2 = plt.plot(col1, col0)
plt.xlabel('col1 data')
plt.ylabel('col2 data')
plt.title('line plot of col1 vs col0')

plt.subplot(3,2,3)
plt.scatter(col2, col1)
plt.xlabel('col2 data')
plt.ylabel('col1 data')
plt.title('scatter plot of col1 vs col0')

plt.subplot(3,2,4)
plt.plot(col2, col1)
plt.xlabel('col2 data')
plt.ylabel('col1 data')
plt.title('line plot of col1 vs col0')

plt.subplot(3,2,5)
plt.scatter(col2, col0)
plt.xlabel('col2 data')
plt.ylabel('col0 data')
plt.title('scatter plot of col2 vs col0')

plt.subplot(3,2,6)
plt.plot(col2, col0)
plt.xlabel('col2 data')
plt.ylabel('col0 data')
plt.title('line plot of col2 vs col0')

plt.subplots_adjust(right = 2, top = 2, wspace = .2, hspace = .5)
```


### 2c) Analysis Question
Which pairs of columns are easier to read as scatter plots and which are easier to read as line graphs?  Why?

When plotting column 1 or column 2 vs column 0, it is easier to read the line plots. These plots shows a continuous progressive change in the data, which is easier to see in the line plot. In the plot of column 2 vs column 0, it is especially easier to read the line graph because there is no visible pattern or correlation in the scatter plot. For plotting column 1 vs column 2, it is easier to read the scatter plot. The line plot is messy because there is no progression in the data as we move along the x axis. The scatter plot, thus, shows the distribution of the data on the graph easier.
